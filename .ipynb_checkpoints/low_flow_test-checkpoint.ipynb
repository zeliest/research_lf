```python
class IFLowFlow(ImpactFunc):
    """Impact functions for tropical cyclones."""

    def __init__(self):
        ImpactFunc.__init__(self)
        self.haz_type = 'LF'
        self.intensity_unit = 'days'

    def get_haz(file,bbox,percentile):
        haz = LowFlow()data
        data_isimip_dir = '/home/zelie/Documents/low_flow/data/hazard/'
        haz.set_from_nc(percentile = percentile, input_dir=data_isimip_dir, yearrange = [1980,2010],yearrange_ref= [1980,2010],bbox = bbox, gh_model='h08',\
                    cl_model='gfdl-esm2m', scenario='historical', soc='histsoc',keep_dis_data = True, min_days_per_month=15)
        return haz

    def get_nc_data(file,bbox):
        data_isimip_dir = '/home/zelie/Documents/low_flow/data/hazard/'

        nc_data = _read_and_combine_nc([1980,2010],data_isimip_dir,'h08','gfdl-esm2m', 'historical', 'histsoc','co2_dis_global_daily',bbox = bbox)

        return nc_data        self.continent = 


    def get_threshold(file,bbox,percentile):
        threshold_grid = _compute_threshold_grid(percentile, [1980,2010], \
                         data_isimip_dir, 'h08', 'gfdl-esm2m', \
                         'historical', 'histsoc','co2_dis_global_daily', bbox)
        return threshold_grid

    def get_dis_day(file,bbox,percentile):

        nc_data = get_nc_data(bbox)
        haz = get_haz(file,bbox,percentile)

        data = nc_data - get_threshold(file,bbox,percentile)
        data = nc_data.where(data<0)
        data = data.resample(time='1M').mean(skipna = True)
        data = _xarray_to_geopandas(data.dropna('time','all'))


        data['Discharge'] = np.nan
        data['Normal Discharge'] = np.nan
        data['Relativ Discharge'] = np.nan


        for i in data.index:

            data['Discharge'][i] = data['dis'][(data["lat"]\
                                    == data['lat'][i]) & (data["lon"] == data['lon'][i])\
                                    & (data["time"] == data['time'][i])].mean(skipna=True)
            data['Normal Discharge'][i] = nc_data['dis'].where((nc_data.coords["lat"] ==\
                                    data['lat'][i])&(nc_data.coords["lon"] == data['lon'][i])).mean(skipna=True)

        data['Relativ Discharge']= data['Discharge']/data['Normal Discharge']

        haz.data['Relativ Discharge'] = np.zeros(len(haz.data))
        for i in haz.data.index:
            haz.data['Relativ Discharge'][i] =  data['Relativ Discharge'][(data['lat']== haz.data['lat'][i])\
                                                                                       & (data['lon']== haz.data['lon'][i])\
                                                                                       & (data['time']== haz.data['time'][i])]

        dis_ndays = haz.data.groupby(['lat','lon','cluster_id'])['dis'].agg('sum').reset_index()
        dis_ndays['Relativ Discharge'] = np.zeros(len(dis_ndays))
        dis_ndays['Relativ Discharge'] =  haz.data.groupby(['lat','lon','cluster_id'])\
        ['Relativ Discharge'].agg('mean').reset_index()['Relativ Discharge']
        dis_ndays = dis_ndays.rename(columns={"dis": "ndays"})

        return dis_ndays
```

```python

# get a data file containing both the number of days and the discharge
def get_dis_day(percentile, min_days_per_month, input_dir, gh_model, cl_model, \
                         bbox, fn_str_var= 'co2_dis_global_daily',scenario = 'historical',
                         soc = 'histsoc', yearrange=[1981,2005], yearrange_ref=[1981,2005]):
    haz = LowFlow()

    haz.set_from_nc(percentile=percentile, input_dir=input_dir, yearrange=yearrange,yearrange_ref=yearrange_ref,bbox = bbox, gh_model=gh_model,\
                    cl_model=cl_model, scenario=scenario, soc=soc, keep_dis_data = True, min_days_per_month=min_days_per_month)
    nc_data = _read_and_combine_nc(yearrange, input_dir, gh_model, cl_model, \
                         scenario, soc, fn_str_var, bbox)
    threshold_grid = _compute_threshold_grid(percentile, yearrange_ref, input_dir, gh_model, cl_model, \
                         scenario, soc, fn_str_var, bbox)
    data = nc_data - threshold_grid
    data = nc_data.where(data<0)
    data = data.resample(time='1M').mean(skipna = True)
    data = _xarray_to_geopandas(data.dropna('time','all'))


    data['Discharge'] = np.nan
    data['Normal Discharge'] = np.nan
    data['Relativ Discharge'] = np.nan


    for i in data.index:

        data['Discharge'][i] = data['dis'][(data["lat"]\
                                == data['lat'][i]) & (data["lon"] == data['lon'][i])\
                                & (data["time"] == data['time'][i])].mean(skipna=True)
        data['Normal Discharge'][i] = nc_data['dis'].where((nc_data.coords["lat"] ==\
                                data['lat'][i])&(nc_data.coords["lon"] == data['lon'][i])).mean(skipna=True)

    data['Relativ Discharge']= data['Discharge']/data['Normal Discharge']

    haz.data['Relativ Discharge'] = np.zeros(len(haz.data))
    for i in haz.data.index:
        haz.data['Relativ Discharge'][i] =  data['Relativ Discharge'][(data['lat']== haz.data['lat'][i])\
                                                                                   & (data['lon']== haz.data['lon'][i])\
                                                                                   & (data['time']== haz.data['time'][i])]

    dis_ndays = haz.data.groupby(['lat','lon','cluster_id'])['dis'].agg('sum').reset_index()
    dis_ndays['Relativ Discharge'] = np.zeros(len(dis_ndays))
    dis_ndays['Relativ Discharge'] =  haz.data.groupby(['lat','lon','cluster_id'])\
    ['Relativ Discharge'].agg('mean').reset_index()['Relativ Discharge']
    dis_ndays = dis_ndays.rename(columns={"dis": "ndays"})

    return dis_ndays

# given a file with yearly flow reduction and corresponding impacts, find the number of days of the event
# and produce impact functions
def make_IF(IF_data, min_days_per_month, percentile, input_dir, gh_model, cl_model, \
                        fn_str_var= 'co2_dis_global_daily',scenario = 'historical',
                         soc = 'histsoc', yearrange=[1981,2005], yearrange_ref=[1981,2005]):
   
    ndays_list = []

    for index, row in IF_data.iterrows():
 
        bbox = literal_eval(IF_data['bbox'].iloc[index])

        discharge_reduction = IF_data['discharge reduction'].iloc[index]
        loss = IF_data['loss'].iloc[index]
        
        dis_ndays = get_dis_day(percentile=percentile, min_days_per_month=min_days_per_month,\
                             input_dir=input_dir, gh_model=gh_model, cl_model=cl_model, \
                             bbox=bbox, fn_str_var=fn_str_var,scenario =scenario,
                             soc=soc, yearrange=yearrange, yearrange_ref=yearrange_ref)
        
        ndays = discharge_reduction/(1-dis_ndays['Relativ Discharge'].mean())*365
        ndays_list.append(ndays)
        
    IF_data[" ".join(['ndays','gh_model','cl_model',])] = ndays_list
    return(IF_data)
        
```

```python
wri_power_plants_path = '/home/zelie/Documents/low_flow/data/exposure/global_power_plant_database.csv'


def load_exposure(bbox=[-15, 35, 40, 70], exp_path=wri_power_plants_path, primary_fuel='Nuclear'):
    """load exposure data from CSV
    Here, we use WRI power plant data from
    https://www.wri.org/publication/global-power-plant-database"""
    exp_df = pd.read_csv(exp_path)
    if primary_fuel:
        exp_df = exp_df[exp_df.primary_fuel==primary_fuel]
    # europe:
    exp_df = exp_df[exp_df.longitude>bbox[0]]
    exp_df = exp_df[exp_df.longitude<bbox[2]]
    exp_df = exp_df[exp_df.latitude>bbox[1]]
    exp_df = exp_df[exp_df.latitude<bbox[3]]
    exp_df['gen_mean']=exp_df[["generation_gwh_2013", "generation_gwh_2014", "generation_gwh_2015", "generation_gwh_2016", "generation_gwh_2017"]].mean(axis=1)
    exp_df['value']=exp_df[["gen_mean", "estimated_generation_gwh"]].mean(axis=1)
    # exp_df['value'] = exp_df.capacity_mw # provide value in MegaWatt
    
    #exp_df['latitude'] = 
    #exp_df['longitude'] = 
    exp_df['if_'] = 1
    exp_df['if_LF'] = 1
    # exp_df = exp_df.drop(columns=['country', 'name', 'gppd_idnr',
    #    'latitude', 'longitude', 'primary_fuel', 'owner', 'source', 'url',
    #    'geolocation_source', 'wepp_id', 'year_of_capacity_data',
    #    'generation_gwh_2013', 'generation_gwh_2014', 'generation_gwh_2015',
    #    'generation_gwh_2016', 'generation_gwh_2017',
    #    'estimated_generation_gwh', 'value', 'value_unit', 'if_TC', 'geometry'])
    

    # Generate Exposures from exposure-DataFranme
    exp_df = Exposures(exp_df)
    exp_df.value_unit = 'gWh'
    exp_df.set_geometry_points() # set geometry attribute (shapely Points) from GeoDataFrame from latitude and longitude
    print('\n' + '\x1b[1;03;30;30m' + 'check method logs:' + '\x1b[0m')
    exp_df.check() # puts metadata that has not been assigned
    return exp_df
```

```python
x0 = np.linspace(0,365,365)
def init_IFS():
    """define impact function imp_fun and return impact function set (IFS)"""
    imp_fun = ImpactFunc()
    imp_fun.haz_type = 'LF'
    imp_fun.id = 1
    imp_fun.name = 'Low flow business interruption'
    imp_fun.intensity_unit = 'days'
    imp_fun.intensity = x0
    imp_fun.mdd = fit_thermo
 # loss to annual operation
    imp_fun.paa = imp_fun.intensity * 0 + 1
    # imp_fun.paa[imp_fun.intensity<mdm] = 0
    imp_fun.check()
    IFS = ImpactFuncSet()
    IFS.append(imp_fun)
    return IFS

IFS_lf = init_IFS()
```
